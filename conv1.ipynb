```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
%matplotlib inline
```

```python
from PIL import Image
img = Image.open('/content/Kakis.jpg')
```

```python
img_mat = np.array(img)
img
```

```python
import torch
from torch.nn.functional import conv2d
```

```python
img_ten = torch.tensor([img_mat], dtype=torch.float)
img_ten.size()
```

```python
img_ten = img_ten.permute(0, 3, 1, 2)
img_ten.size()
```

```python
sob = [[1, 1, 1],
       [-1, 0, 1],
       [-1, -1,-1]]
'''sob = [[0, 5, 0],
       [0, 1, 0],
       [0, -5, 0]]'''
ker = [[sob, sob, sob]]
ker = torch.tensor(ker, dtype=torch.float)
```

```python
for i in range(1000):
  img_cov = conv2d(img_ten, ker)
img_cov = img_cov.permute(0, 2, 3, 1)
```

```python
plt.imshow(img_cov[0, :, :, 0], cmap = cm.Blues)
plt.show()
```
